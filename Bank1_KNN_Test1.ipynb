```python
dataset = pd.read_csv('bank-data.csv', sep = ';', header = 0)
```

```python
dataset_w = dataset[["age", "marital", "default", "housing", "loan", "y"]]
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=14)
```

```python
dataset_w['marital'].replace(['single', 'married','divorced', 'unknown'], [0, 1, 3, 99], inplace= True)
dataset_w['loan'].replace(['no', 'yes', 'unknown'], [0, 1, 99], inplace= True)
dataset_w['y'].replace(['no', 'yes'], [0, 1], inplace= True)
dataset_w['housing'].replace(['no', 'yes', 'unknown'], [0, 1, 99], inplace= True)
dataset_w['default'].replace(['no', 'yes', 'unknown'], [0, 1, 99], inplace= True)
```

```python
dataset_w.head()
```

```python
y = dataset_w['y']
X = dataset_w.drop('y', axis=1)
```

```python
model.fit(X, y)
```

```python
model.score(X, y)
```

```python
def epargne(model, age= 32, marital = 3, default = 2, housing = 0, loan = 0):
  x = np.array([age, marital, default, housing, loan]).reshape(1, 5)
  print(model.predict(x))
  print(model.predict_proba(x))
```
